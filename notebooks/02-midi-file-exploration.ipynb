# Midi file exploration

Lets pick a random song from the dataset

```python
import os
import random

f = []

for root, dirs, files in os.walk("../data/raw"):
    for file in files:
        if file.endswith(".mid"):
             f.append(os.path.join(root, file))

example = random.choice(f)
example
```

We will use music21 library for operating with the midi file. With this library we can import the midi file and print the music sheet:

```python
from music21 import converter, instrument, note, chord

midi = converter.parse(example)
midi.show()
```

Also its possible to play the file:

```python
midi.show('midi')
```

Or write and read the contents:

```python
midi.show('text')
```
